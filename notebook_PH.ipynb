```python
from pandas import read_csv
```

```python
file = '../data/TCE-PB-Portal-Gestor-Licitacoes_Propostas.txt'
df = read_csv(file,sep='|')
```

```python
df
```

```python
def adiciona_valor_binario(objeto):
    if 'show' in objeto.lower() or 'musica' in objeto.lower():
        return True
    else:
        return False
```

```python
df['show'] = df['objeto_licitacao'].apply(adiciona_valor_binario)
```

```python
df_shows = df.query('show == True')
```
